```python
import cv2 
import matplotlib.pyplot as plt
from scipy import misc

```

```python
cat_im = '../hybrid_images/data/cat.bmp'
dog_im =  '../hybrid_images/data/dog.bmp'

c_read = plt.imread(cat_im)
d_read = plt.imread(dog_im)

plt.imshow(c_read)
plt.show()
plt.imshow(d_read)
plt.show()
```

```python
b_dog =  cv2.blur(d_read,(10,10))
b_cat =  cv2.blur(c_read,(10,10))

plt.imshow(b_dog)
plt.show()
plt.imshow(b_cat)
plt.show()
```

```python
gauss_cat = cv2.GaussianBlur(c_read,(5,5),20)
gauss_dog = cv2.GaussianBlur(d_read ,(5,5),20)

plt.imshow(gauss_cat)
plt.show()
plt.imshow(gauss_dog)
plt.show()
```

```python
hi_cat = cv2.subtract( c_read,gauss_cat )
hi_dog = cv2.subtract( d_read, gauss_dog )
plt.imshow(hi_cat)
plt.show()
plt.imshow(hi_dog)
plt.show()
```

```python
hyb_catdog = cv2.add(gauss_dog,hi_cat)
plt.imshow(hyb_catdog)
plt.show()
```

```python
u_hyb_catdog = cv2.pyrDown(hyb_catdog)
plt.imshow(u_hyb_catdog)
plt.show()
```

```python
uu_hyb_catdog = cv2.pyrDown(u_hyb_catdog)
plt.imshow(uu_hyb_catdog)
plt.show()
```

```python
uuu_hyb_catdog = cv2.pyrDown(uu_hyb_catdog)
plt.imshow(uuu_hyb_catdog)
plt.show()
```
