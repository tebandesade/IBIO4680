```python
import cyvlfeat as vlf
import cv2
import os
```

```python
tImagesdir = 'datalab9/TrainImages/'
```

```python
for cat in os.listdir(tImagesdir):
    path = os.path.join(tImagesdir,cat)
    for im in os.listdir(path):
        subpa = os.path.join(path,im)
        img = cv2.imread(subpa)
        hog = cv2.HOGDescriptor()
        ms  = hog.detectMultiScale(img)
        print(ms)
```
